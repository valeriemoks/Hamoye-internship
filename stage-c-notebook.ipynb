```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
import lightgbm as lgb
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from xgboost import XGBClassifier
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.model_selection import RandomizedSearchCV


```

```python
df=pd.read_csv("/kaggle/input/electrical/Data_for_UCI_named.csv")
df.head()
```

```python
df.shape
df.info()
```

```python
df.isnull().sum()
```

```python
df= df.dropna()
df.isnull().sum()
```

```python
encoder = LabelEncoder()
# Prepare the feature and target variables
X = df.drop(columns=["stab", "stabf"])
y = df["stabf"]

# Encode the target variable
y_encoded = encoder.fit_transform(y)

# Split the data into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.3, random_state=0)

# Create a random forest classifier object
rf = RandomForestClassifier(random_state=1)

# Train the random forest classifier on the training data
rf.fit(x_train, y_train)

# Print the accuracy on the test set
print(f"Accuracy on test set: {round(rf.score(x_test, y_test), 4)}")
```

```python
# Create the XGBoost classifier object
xgboost = XGBClassifier(random_state=1)

# Train the classifier on the training data
xgboost.fit(x_train, y_train)

# Make predictions on the test set
y_pred = xgboost.predict(x_test)

# Calculate the accuracy score
accuracy = accuracy_score(y_test, y_pred)

# Print the accuracy on the test set
print(f"Accuracy on test set: {round(accuracy, 4)}")
```

```python
# Create the LightGBM classifier object
lgb_classifier = lgb.LGBMClassifier(random_state=1)

# Train the classifier on the training data
lgb_classifier.fit(x_train, y_train)

# Make predictions on the test set
y_pred = lgb_classifier.predict(x_test)

# Calculate the accuracy score
accuracy = accuracy_score(y_test, y_pred)

# Print the accuracy on the test set
print(f"Accuracy on test set: {round(accuracy, 4)}")
```

```python
xtree = ExtraTreesClassifier(random_state=1)
xtree.fit(x_train, y_train)
```

```python
xtree = ExtraTreesClassifier(random_state=1)
xtree.fit(x_train, y_train)
param_grid1 = {
    "n_estimators": [100, 300, 500, 1000],
    "min_samples_split": [2, 5, 7, 10],
    "min_samples_leaf": [4, 6, 8, 16],
    "max_features": ["auto", "log2", None]
}
cv = RandomizedSearchCV(estimator=xtree, param_distributions=param_grid1, cv=5, n_iter=10, scoring="accuracy", n_jobs=-1, verbose=1, random_state=1)
cv.fit(x_train, y_train)
print(cv.best_params_)
```

```python
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.model_selection import RandomizedSearchCV

# Create the ExtraTreesClassifier object
xtree = ExtraTreesClassifier(random_state=1)

# Fit the ExtraTreesClassifier on the training data
xtree.fit(x_train, y_train)

# Define the parameter grid for randomized search
param_grid1 = {
    "n_estimators": [100, 300, 500, 1000],
    "min_samples_split": [2, 5, 7, 10],
    "min_samples_leaf": [4, 6, 8, 16],
    "max_features": ["auto", "log2", None]
}

# Perform randomized search cross-validation
cv = RandomizedSearchCV(estimator=xtree, param_distributions=param_grid1, cv=5, n_iter=10, scoring="accuracy", n_jobs=-1, verbose=1, random_state=1)
cv.fit(x_train, y_train)

# Print the best hyperparameters
print(cv.best_params_)
```

```python
print(f"Accuracy on test set: {round(xtree.score(x_test, y_test), 4)}")
```

```python
xtree2 = ExtraTreesClassifier(n_estimators = 1000 , min_samples_split = 2 , min_samples_leaf = 8, max_features = None, random_state=1)
xtree2.fit(x_train, y_train)
print(f"Accuracy on test set: {round(xtree2.score(x_test, y_test), 4)}")
     
```

```python
# Get the best estimator from the randomized search
best_estimator = cv.best_estimator_

# Fit the best estimator on the training data
best_estimator.fit(x_train, y_train)

# Get the feature importances
feature_importances = best_estimator.feature_importances_

# Create a dictionary of feature importance scores
importance_scores = {feature: importance for feature, importance in zip(X.columns, feature_importances)}

# Sort the importance scores in descending order
sorted_scores = sorted(importance_scores.items(), key=lambda x: x[1], reverse=True)

# Print the most and least important features
most_important = sorted_scores[0]
least_important = sorted_scores[-1]

print("Most important feature:", most_important[0])
print("Least important feature:", least_important[0])
```
